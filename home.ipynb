```python
from artwork import Artwork
from location import Location
from visitor import Visitor
from ticket import Ticket
from exhibition import Exhibition
from special_event import SpecialEvent

# Create an instance of the Artwork class
artwork = Artwork(id=1, title="Mona Lisa", artist="Leonardo da Vinci", creation_date="1503–1506",
                  historical_significance="Iconic masterpiece", location="Louvre Museum")

# Print information about the artwork
print("Artwork Information:")
print("ID:", artwork.get_id())
print("Title:", artwork.get_title())
print("Artist:", artwork.get_artist())
print("Creation Date:", artwork.get_creation_date())
print("Historical Significance:", artwork.get_historical_significance())
print("Location:", artwork.get_location())

# Create an instance of the Location class
location = Location(id=1, type="Museum", name="Louvre Museum")

# Print information about the location
print("\nLocation Information:")
print("ID:", location.get_id())
print("Type:", location.get_type())
print("Name:", location.get_name())

# Create an instance of the Visitor class
visitor = Visitor(id=1, name="Alyaa", age=30, category="Adult", national_id="1234567890")

# Print information about the visitor
print("\nVisitor Information:")
print("ID:", visitor.get_id())
print("Name:", visitor.get_name())
print("Age:", visitor.get_age())
print("Category:", visitor.get_category())
print("National ID:", visitor.get_national_id())

# Create an instance of the Ticket class
ticket = Ticket(id=1, type="General", date="2024-03-17", price=20.00)

# Print information about the ticket
print("\nTicket Information:")
print("ID:", ticket.get_id())
print("Type:", ticket.get_type())
print("Date:", ticket.get_date())
print("Price:", ticket.get_price())

# Create an instance of the Exhibition class
exhibition = Exhibition(id=1, title="Renaissance Art Exhibition", start_date="2024-04-01", 
                        end_date="2024-06-30", location="Louvre Museum")

# Print information about the exhibition
print("\nExhibition Information:")
print("ID:", exhibition.get_id())
print("Title:", exhibition.get_title())
print("Start Date:", exhibition.get_start_date())
print("End Date:", exhibition.get_end_date())
print("Location:", exhibition.get_location())

# Create an instance of the SpecialEvent class
special_event = SpecialEvent(id=1, title="Concert in the Garden", date="2024-05-15", location="Louvre Garden", 
                             ticket_price=50.00)

# Print information about the special event
print("\nSpecial Event Information:")
print("ID:", special_event.get_id())
print("Title:", special_event.get_title())
print("Date:", special_event.get_date())
print("Location:", special_event.get_location())
print("Ticket Price:", special_event.get_ticket_price())
```
